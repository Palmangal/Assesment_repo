```python
import requests
import matplotlib.pyplot as plt
```

```python
response = requests.get("https://data.nasa.gov/resource/y77d-th95.json")
data = response.json()
```

```python
earth_meteorites_before_2000 = []
earth_meteorites_coordinates_before_1970 = []
earth_meteorites_mass_above_10000kg = []
```

```python
# Get all the Earth meteorites that fell before the year 2000
# Get all the earth meteorites co-ordinates who fell before the year 1970
# Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg
```

```python
# Iterate over the meteorite data and extract the required information
for meteorite in data:
    if meteorite.get('reclat') and meteorite.get('reclong'):  # Check if coordinates are available
        year = int(meteorite.get('year', '0')[:4])  # Extract the year from the 'year' field
        
        if year < 2000:  # Earth meteorites that fell before the year 2000
            earth_meteorites_before_2000.append(meteorite)
        
        if year < 1970:  # Earth meteorites that fell before the year 1970
            coordinates = (float(meteorite['reclat']), float(meteorite['reclong']))
            earth_meteorites_coordinates_before_1970.append(coordinates)
        
        mass = float(meteorite.get('mass', '0'))  # Extract the mass of the meteorite
        if mass > 10000:  # Earth meteorites with mass greater than 10000kg
            earth_meteorites_mass_above_10000kg.append(meteorite)

# Print the results
print("Earth meteorites that fell before the year 2000:")
for meteorite in earth_meteorites_before_2000:
    print(meteorite['name'])

print("\nEarth meteorite coordinates before the year 1970:")
for coordinates in earth_meteorites_coordinates_before_1970:
    print(coordinates)

print("\nEarth meteorites with mass greater than 10000kg:")
for meteorite in earth_meteorites_mass_above_10000kg:
    print(meteorite['name'], meteorite['mass'])
```

```python
# Plotting the distribution of meteorite masses
masses = [float(meteorite.get('mass', '0')) for meteorite in data if meteorite.get('mass')]
plt.hist(masses, bins=50)
plt.xlabel('Mass (kg)')
plt.ylabel('Frequency')
plt.title('Distribution of Meteorite Masses')
plt.show()
```
